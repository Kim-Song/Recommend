```python
import pandas as pd
import numpy as np
import os

data_path = '../DataSet/'

user_baekjoon_df = pd.read_csv(os.path.join(data_path, 'Baekjoon_유저_크롤링.csv'))
problem_baekjoon_df = pd.read_csv(os.path.join(data_path, 'Baekjoon_문제_크롤링.csv'))
user_solvedac_df = pd.read_csv(os.path.join(data_path, 'solved.ac_유저_크롤링.csv'))
```

# Final_Problem_Dataset 만들기

```python
problem_baekjoon_df
```

```python
problem_baekjoon_df.info()
```

```python
# 알고리즘 빈도수가 낮은 것만 존재하는 문제의 경우, 삭제하자.

import ast

#문자열로 표현된 리스트를 리스트형으로 변환.
problem_baekjoon_df['problem_algorithm'] = problem_baekjoon_df['problem_algorithm'].apply(ast.literal_eval)

from collections import Counter

s = []
for p in problem_baekjoon_df['problem_algorithm']:
    for i in p:
        s.append(i)
C = Counter(s)
C_dict = sorted(C.items(), key=lambda x: x[1])
C_dict
```

```python
del_list = [x[0] for x in C_dict if x[1] < 10]
del_list
```

```python
problem_baekjoon_df = problem_baekjoon_df[~problem_baekjoon_df['problem_algorithm'].apply(lambda x: all(item in x for item in del_list))]
problem_baekjoon_df['cat_problem_avg_try'] = problem_baekjoon_df['problem_avg_try'].apply(lambda x: int(x - 0.5) + 1 if x > 0.5 else 1)
```

```python
#삭제된 것이 없음 즉, 알고리즘이 하나만 뜨는 것이라도 다른 알고리즘이 곁들어져서 큰 문제가 안됨.
problem_baekjoon_df
```

```python
problem_baekjoon_df.to_csv(os.path.join(data_path, 'final_Baekjoon_문제_크롤링.csv'), index=False)
```

# Final_User_Dataset 만들기

```python
user_baekjoon_df
```

```python
user_baekjoon_df.info()
```

```python
# 유저 아이디 목록이 존재하지 않는 것 존재 -> 데이터 오류, 삭제 요망

user_baekjoon_df[user_baekjoon_df['user_id_list']!=user_baekjoon_df['user_id_list']]
```

```python
#user_id, code_language에 Nan값 제거, code_length는 의미 없는 feature -> 제거
user_baekjoon_df = user_baekjoon_df[user_baekjoon_df['user_id_list'] == user_baekjoon_df['user_id_list']]
user_baekjoon_df = user_baekjoon_df[user_baekjoon_df['code_language_list'] == user_baekjoon_df['code_language_list']]
```

```python
user_baekjoon_df
```

```python
# Problem number 형 변환
# submit_time 데이터형 변환
# memory, time은 틀렸을 시 nan값 -> 결측치는 아니므로 -1 로 대체

user_baekjoon_df['problem_number_list'] = user_baekjoon_df['problem_number_list'].astype(int)
user_baekjoon_df['code_memory_list'].fillna(-1, inplace=True)
user_baekjoon_df['code_time_list'].fillna(-1, inplace=True)
user_baekjoon_df['code_memory_list'] = user_baekjoon_df['code_memory_list'].astype(int)
user_baekjoon_df['code_time_list'] = user_baekjoon_df['code_time_list'].astype(int)

user_baekjoon_df
```

```python
#크롤링한 문제 목록에 없는 문제에 대한 결과들을 제외.

problem_list = list(problem_baekjoon_df['problem_number'])

user_baekjoon_df = user_baekjoon_df[user_baekjoon_df['problem_number_list'].isin(problem_list)]
user_baekjoon_df
```

```python
#크롤링한 유저 목록에 없는 유저에 대한 결과들은 제외.
user_list = list(user_solvedac_df['user_name'])

user_baekjoon_df = user_baekjoon_df[user_baekjoon_df['user_id_list'].isin(user_list)]
user_baekjoon_df
```

```python
#유저 티어 추가.

user_baekjoon_df = pd.merge(user_baekjoon_df, user_solvedac_df[['user_name', 'user_tier']], how='inner', left_on='user_id_list', right_on = 'user_name')
user_baekjoon_df
```

```python
user_baekjoon_df.isnull().sum()
```

```python
from collections import Counter

C = Counter(list(user_baekjoon_df['problem_result_list']))
C_dict = sorted(C.items(), key=lambda x: -x[1])
C_dict
```

```python
user_baekjoon_df.columns = ['submit_number', 'user_id', 'problem_number', 'problem_result', 'code_memory','code_time', 'code_language', 'user_tier']
```

```python
user_baekjoon_df.to_csv(os.path.join(data_path, 'final_Baekjoon_유저_크롤링.csv'),index=False)
```

# Encoder 생성

```python
from sklearn.preprocessing import LabelEncoder
import joblib
from collections import Counter
import ast

encoder_data_path = '../DataSet/Encoder/'

user_list = list(user_solvedac_df['user_name'])
problem_list = list(problem_baekjoon_df['problem_number'])
#problem_baekjoon_df['problem_algorithm'] = problem_baekjoon_df['problem_algorithm'].apply(ast.literal_eval)

s = []
for p in problem_baekjoon_df['problem_algorithm']:
    for i in p:
        s.append(i)

C = Counter(s)
algorithm_list = list(C.keys())


user_ID_Encoder = LabelEncoder()
problem_number_Encoder = LabelEncoder()
problem_algorithm_Encoder = LabelEncoder()

user_ID_Encoder.fit(user_list)
problem_number_Encoder.fit(problem_list)
problem_algorithm_Encoder.fit(algorithm_list)

joblib.dump(user_ID_Encoder, os.path.join(encoder_data_path, 'user_ID_Encoder.joblib'))
joblib.dump(problem_number_Encoder, os.path.join(encoder_data_path, 'problem_number_Encoder.joblib'))
joblib.dump(problem_algorithm_Encoder, os.path.join(encoder_data_path, 'problem_algorithm_Encoder.joblib'))
```
