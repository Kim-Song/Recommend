```python
import pandas as pd
import numpy as np
import time
import os
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import pickle
import re

from selenium.common.exceptions import StaleElementReferenceException

chrome_options = Options()
chrome_options.add_argument('--headless')  # 웹 브라우저를 표시하지 않고 백그라운드에서 실행
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
tier_dict = {
    'Silver V': 6,
    'Silver IV': 7,
    'Silver III': 8,
    'Silver II': 9,
    'Silver I': 10,
    'Gold V': 11,
    'Gold IV': 12,
    'Gold III': 13,
    'Gold II': 14,
    'Gold I': 15,
    'Platinum V': 16,
    'Platinum IV': 17,
    'Platinum III': 18,
    'Platinum II': 19,
    'Platinum I': 20,
    'Diamond V': 21,
    'Diamond IV': 22
}
```

```python


total_user_name_list = []
total_user_tier_list = []
total_user_spc_list = []

for t in range(20,702):

    solved_user_url = 'https://solved.ac/ranking/tier?page=' + str(t)
    print(solved_user_url)
    driver = webdriver.Chrome('chromedriver.exe', options=chrome_options)
    driver.get(solved_user_url)

    #닉네임 리스트 얻는 코드
    user_name_object = driver.find_elements_by_css_selector('.css-1wnvjz2')
    user_tier_object = driver.find_elements_by_css_selector('.css-1vnxcg0')
    #푼 문제 수 얻는 코드
    user_spc_object = driver.find_elements_by_css_selector('.css-1ujcjo0')
    
    sub_name_list = []
    sub_tier_list = []
    sub_spc_list = []

    for uno in user_name_object:
        sub_name_list.append(uno.text.strip())
    
    for uto in user_tier_object:
        ut = uto.get_attribute('alt')
        sub_tier_list.append(tier_dict[ut])
    
    for i in range(len(user_name_object)):
        sub_spc_list.append(user_spc_object[2*i + 1].text)


    total_user_name_list.extend(sub_name_list)
    total_user_tier_list.extend(sub_tier_list)
    total_user_spc_list.extend(sub_spc_list)


    lists_to_save = {
        'total_user_name_list': total_user_name_list,
        'total_user_tier_list': total_user_tier_list,
        'total_user_spc_list': total_user_spc_list
    }

    file_name = "solved_user_lists.pkl"
    with open(file_name, 'wb') as file:
        pickle.dump(lists_to_save, file) 
    
    driver.quit()

```

```python
file_name = "solved_user_lists.pkl"

with open(file_name, 'rb') as file:
    loaded_dicts = pickle.load(file)

total_user_name_list = loaded_dicts['total_user_name_list']
total_user_tier_list = loaded_dicts['total_user_tier_list']
total_user_spc_list = loaded_dicts['total_user_spc_list']
```
